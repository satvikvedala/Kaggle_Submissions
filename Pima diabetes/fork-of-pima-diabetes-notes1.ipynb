```python
cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age']
```

```python
arr = []
for i in range(len(cols)):
    temp = []
    for j in range(len(cols)):
        if i!=j:
            cr = df[cols[i]].corr(df[cols[j]])
            temp.append(cr)
        if i == j:
            temp.append(1)
    arr.append(temp)
print(arr)
            
```

```python
for j in range(len(arr)):
    ma = 0
    index = 0
    for i in range(len(arr[j])):
        if arr[j][i]!=1 and arr[j][i]>ma:
            ma = arr[j][i]
            index = i
    print(cols[j]," ",cols[index])
```

```python
arr = []
for i in range(len(df['Glucose'])):
    if df['Glucose'][i] == 0:
        arr.append(df['Age'][i])
print(arr)
```

```python
#for 0-10 years
arr_1_to_10 = []
arr_11_to_20 = []
arr_21_to_30 = []
arr_31_to_40 = []
arr_41_to_50 = []
arr_51_to_60 = []
arr_61_to_70 = []
arr_71_to_80 = []
arr_81_to_90 = []
arr_91_to_100 = []
```

```python
for i in range(len(df['Age'])):
    num = df['Age'][i]//10
    if num == 0:
        arr_1_to_10.append(df['Glucose'][i])
    elif num == 1:
        arr_11_to_20.append(df['Glucose'][i])
    elif num == 2:
        arr_21_to_30.append(df['Glucose'][i])
    elif num == 3:
        arr_31_to_40.append(df['Glucose'][i])
    elif num == 4:
        arr_41_to_50.append(df['Glucose'][i])
    elif num == 5:
        arr_51_to_60.append(df['Glucose'][i])
    elif num == 6:
        arr_61_to_70.append(df['Glucose'][i])
    elif num == 7:
        arr_71_to_80.append(df['Glucose'][i])
    elif num == 8:
        arr_81_to_90.append(df['Glucose'][i])
    elif num == 9:
        arr_91_to_100.append(df['Glucose'][i])
    else:
        arr_91_to_100.append(df['Glucose'][i])
Glucose_array = [0,0,sum(arr_21_to_30)//len(arr_21_to_30),
                 sum(arr_31_to_40)//len(arr_31_to_40),
                 sum(arr_41_to_50)//len(arr_41_to_50),
                 sum(arr_51_to_60)//len(arr_51_to_60),
                 sum(arr_61_to_70)//len(arr_61_to_70),
                 sum(arr_71_to_80)//len(arr_71_to_80),
                 sum(arr_81_to_90)//len(arr_81_to_90),0]
#removed arr_1_to_10,arr_11_to_20 and arrarr_91_to_100. As their lengths are 0
print(Glucose_array)
for i in range(len(df['Glucose'])):
    if df['Glucose'][i] == 0:
        num = df['Age'][i]//10
        df['Glucose'][i].replace(to_replace=0,value = Glucose_array[num])
count_zero = 0
for i in range(len(df['Glucose'])):
    if df['Glucose'][i] == 0:
        count_zero+=1
print(count_zero)

    
```

```python
Y = df['Outcome']
X = df.drop('Outcome',axis = 1)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size = 0.3,random_state = 0)
```

```python
ytrain
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit_transform(xtrain)
ss.transform(xtest)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver='liblinear')
```

```python
lr.fit(xtrain,ytrain)
```

```python
y_pred = lr.predict(xtest)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,y_pred))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,ytest))
```
